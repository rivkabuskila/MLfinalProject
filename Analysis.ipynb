```python
# importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#choose a data type mta or boa
name1 = 'mta'
name2 = 'boa'
file_path2 = f'{name1}.csv'

# קריאת הנתונים משני הקבצים
data = pd.read_csv(file_path2)
X = data.iloc[:, 0:19].values
```
